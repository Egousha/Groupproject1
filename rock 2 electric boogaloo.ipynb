```python
# import random to use randrange in computer function
import random

#Function that evaluates the computer's choice
def computerchoice():
    #generates random numbers 0,1,or 2
    choice = random.randrange(2) 
    #assigns the random numbers to rock,paper, or scissor
    if choice == 0: 
        choice = 'Rock' 
    if choice == 1: 
        choice = 'Paper' 
    if choice==2: 
        choice = 'Scissors' 
    return choice 

#Function that evaluates the player's choice
def playerchoice(): 
    #gathers the input from the player, says which number corresponds to which choice
    choice = int(input("Type 0 for Rock, 1 for Paper, or 2 for Scissors: ")) 
    #If somehow you put in the wrong number, it will tell you to put in a correct number
    while choice != 0 and choice != 1 and choice != 2: 
        print('The numbers are rock(0), paper(1),or scissors(2).') 
        choice = int(input('Seriously? Enter a valid number: ')) 
    #assigns the player's input to rock, paper, or scissor
    if choice == 0: 
        choice = 'Rock' 
    if choice == 1: 
        choice = 'Paper' 
    if choice==2: 
        choice = 'Scissors' 
    return choice 

#function that evalues who wins based on each pick
def dub(computerchoice, playerchoice): 
    if computerchoice == playerchoice:
        print("It's a tie")
    if computerchoice == 'Scissors' and playerchoice == 'Rock':
        print('You win')
    if computerchoice == 'Paper' and playerchoice == 'Scissors': 
        print('You win')
    if computerchoice == 'Rock' and playerchoice == 'Paper': 
        print('You win')
    if computerchoice == 'Rock' and playerchoice == 'Scissors':
        print('You lose')
    if computerchoice == 'Scissors' and playerchoice == 'Paper':
        print('You lose')
    if computerchoice == 'Paper' and playerchoice == 'Rock':
        print('You Lose')

#function that evaluates the game itself
def game(): 
    #continue is a function in python, so we spelled it this way
    #Starts at yes so that the game starts normally
    continuue = 'yes'  
    #while loop that takes in the computer and player choice
    while continuue == 'yes':  
        print('Please input the number corresponding to the object you want to choose.') 
        computer=computerchoice()
        player=playerchoice()
        #prints what the computer chose
        print('The Computer chose', computer) 
        #prints what you chose
        print('You chose', player) 
        #Prints the outcome of the winning function
        dub(computer, player) 
        #asks the player if they want to continue, if they say yes, restarts loop, say no, ends loop.
        continuue = input("Would you like to continue? Type 'yes' for yes or 'no' for no. ") 
#runs game function
game() 
```
